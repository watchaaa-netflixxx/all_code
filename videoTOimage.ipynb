```python
# pip install moviepy
import os
import cv2
import moviepy.editor
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
from moviepy.editor import VideoFileClip
   
video_path = './uploaded_video/stu1_43.mp4'     # **************************************** input
output_div_path = './15s_cut_video/'          # ****************************************

img_div_path = './32_imgs'                       # ****************************************
```

```python
video_duration = VideoFileClip(video_path).duration

# Define the clip duration (16 seconds)
clip_duration = 16

# Calculate the start time for the clip
start_time = video_duration / 2 - (clip_duration / 2)

# Extract the 16-second clip
output_path = output_div_path + "15s.mp4"  # Output file name
ffmpeg_extract_subclip(video_path, start_time, start_time + clip_duration, targetname=output_path)
```

```python
def getFrames(vid, output, rate=0.5, frameName='frame'):

    vidcap = cv2.VideoCapture(vid)
    clip = moviepy.editor.VideoFileClip(vid)    # 동영상 파일을 불러오기
    seconds = clip.duration                     # 동영상 전체 길이

    count = 0
    frame = 0

    if not os.path.isdir(output):
        os.mkdir(output)

    success = True
    while success:
        vidcap.set(cv2.CAP_PROP_POS_MSEC,frame*1000)
        success,image = vidcap.read()

        ## Stop when last frame is identified
        # print(frame)
        if frame > seconds or not success:
            break

        # print('extracting frame ' + frameName + '-%s.png' % str(count).zfill(2))
        name = output + '/' + frameName + '-%s.png' % str(count).zfill(2) # save frame as PNG file
        cv2.imwrite(name, image)
        frame += rate
        count += 1
```

```python
def vid_cut(video_path, output_div_path):
    
    # input 예시
    # video_path = './uploaded_video/stu1_43.mp4'     
    # output_div_path = './15s_cut_video/'          

    video_duration = VideoFileClip(video_path).duration

    # Define the clip duration (16 seconds)
    clip_duration = 16

    # Calculate the start time for the clip
    start_time = video_duration / 2 - (clip_duration / 2)

    # Extract the 16-second clip
    output_path = output_div_path + "15s.mp4"  # Output file name
    ffmpeg_extract_subclip(video_path, start_time, start_time + clip_duration, targetname=output_path)

```

```python
def vid2img(video_path, img_div_path, rate=0.5, frameName='frame'):
    # input 예시
    # getFrames(vid=video_path, output=img_div_path, frameName='frame')  
    # video_path = ./15s_cut_video/15s.mp4
    # img_div_path = './32_imgs'

    vidcap = cv2.VideoCapture(video_path)
    clip = moviepy.editor.VideoFileClip(video_path)    # 동영상 파일을 불러오기
    seconds = clip.duration                     # 동영상 전체 길이

    count = 0
    frame = 0

    if not os.path.isdir(img_div_path):
        os.mkdir(img_div_path)

    success = True
    while success:
        vidcap.set(cv2.CAP_PROP_POS_MSEC,frame*1000)
        success,image = vidcap.read()

        ## Stop when last frame is identified
        # print(frame)
        if frame > seconds or not success:
            break

        # print('extracting frame ' + frameName + '-%s.png' % str(count).zfill(2))
        name = img_div_path + '/' + frameName + '-%s.png' % str(count).zfill(2) # save frame as PNG file
        cv2.imwrite(name, image)
        frame += rate
        count += 1
```
