```python
import math
def calculate_3d(point1, point2, point3):
    # Calculate the vectors between the left hand, elbow, and shoulder landmarks
    point_1_2 = [(point2[i] - point1[i]) for i in range(3)]
    point_2_3 = [(point2[i] - point3[i]) for i in range(3)]

    # Calculate the dot product and the magnitudes of the vectors
    dot_product = sum([point_1_2[i] * point_2_3[i] for i in range(3)])
    point_1_2_mag = math.sqrt(sum([coord**2 for coord in point_1_2]))
    point_2_3_mag = math.sqrt(sum([coord**2 for coord in point_2_3]))

    # Calculate the angle between the left hand, elbow, and shoulder landmarks in degrees
    angle = math.degrees(math.acos(dot_product / (point_1_2_mag * point_2_3_mag)))

    return angle
```

```python
point_groups = [
    ('LEFT_SHOULDER', 'LEFT_ELBOW', 'LEFT_WRIST'),
    ('LEFT_HIP', 'LEFT_SHOULDER', 'LEFT_ELBOW'),
    ('LEFT_KNEE', 'LEFT_HIP', 'LEFT_SHOULDER'),
    ('LEFT_ANKLE', 'LEFT_KNEE', 'LEFT_HIP'),
    ('RIGHT_SHOULDER', 'RIGHT_ELBOW', 'RIGHT_WRIST'),
    ('RIGHT_HIP', 'RIGHT_SHOULDER', 'RIGHT_ELBOW'),
    ('RIGHT_KNEE', 'RIGHT_HIP', 'RIGHT_SHOULDER'),
    ('RIGHT_ANKLE', 'RIGHT_KNEE', 'RIGHT_HIP')
]

# 열 이름 리스트 만들기
column_names = [f'{group[1]}_ANGLE' for group in point_groups]

# 각 그룹의 각도 계산 및 데이터프레임에 추가
for group, col_name in zip(point_groups, column_names):
    a[col_name] = a.apply(lambda row: calculate_3d(
        [row[f'{group[0]}.x'], row[f'{group[0]}.y'], row[f'{group[0]}.z']],
        [row[f'{group[1]}.x'], row[f'{group[1]}.y'], row[f'{group[1]}.z']],
        [row[f'{group[2]}.x'], row[f'{group[2]}.y'], row[f'{group[2]}.z']]
    ), axis=1)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
num_classes = len(np.unique(y))
y = keras.utils.to_categorical(y, num_classes)
y.shape
```

```python
# 데이터를 32개의 시퀀스로 묶습니다.
import numpy as np
sequence_length = 32  # 시퀀스 길이 설정
Xsequence, ysequence = [], []

for i in range(0, len(X) - sequence_length + 1, sequence_length):
    Xsequence.append(X[i:i+sequence_length])
    ysequence.append(y[i+sequence_length-1])

Xsequence = np.array(Xsequence)
ysequence = np.array(ysequence)
```

```python
numbers=list(range(len(Xsequence[0])))
even_numbers=[num for num in numbers if num%2==0]
odd_numbers=[num for num in numbers if num%2!=0]

#
even_numbers.sort()
odd_numbers.sort(reverse=True)

result=even_numbers+odd_numbers
```

```python
# 데이터를 32개의 시퀀스로 나누고 정렬하고 데이터 재배열
def divide_and_sort_sequences(data, sequence_length, sort_column):
    num_sequences = len(data) // sequence_length
    sequence_list = []

    for i in range(num_sequences):
        sequence = data[i * sequence_length:(i + 1) * sequence_length]
        if number==0 or number==1 or number==3:
            sorted_sequence = sequence.sort_values(by=sort_column, ascending=True)
        else:
            sorted_sequence = sequence.sort_values(by=sort_column, ascending=False)
        # 이 부분에서 result 리스트를 활용하여 정렬된 순서로 시퀀스를 재배열
        result2 = []
        for j in range(len(sorted_sequence)):
            result2.append(sorted_sequence.iloc[result[j], :])

        sequence_list.append(result2)

    return sequence_list
```

```python
# 시퀀스로 데이터를 나누고 정렬
sequences = divide_and_sort_sequences(X, sequence_length, sort_column)
```

```python
xsequence=np.array(sequences)
xsequence.shape,ysequence.shape
```

```python
pd.DataFrame(xsequence[0])
```

```python
from sklearn.model_selection import train_test_split

n = Xsequence.shape[0]
# 먼저 train과 test를 분리합니다.
X_train, X_test, y_train, y_test = train_test_split(xsequence, ysequence, test_size=0.1, shuffle=True, random_state=42)
# 다음으로 train을 train과 validation으로 분리합니다.
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, shuffle=True, random_state=42)
X_train.shape, y_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization

sequence_length = 32  # 시퀀스 길이 (설정에 맞게 조정해야 함)
input_dim = 19  # 입력 차원 (설정에 맞게 조정해야 함)

model = tf.keras.models.Sequential([
    LSTM(128, dropout=0.2, input_shape=(sequence_length, input_dim), return_sequences=True),
    BatchNormalization(),
    LSTM(256, return_sequences=True),
    LSTM(512, return_sequences=True),
    Dropout(0.1),
    LSTM(256, return_sequences=True),
    LSTM(128, return_sequences=True),
    LSTM(64, return_sequences=True),
    Dropout(0.1),
    LSTM(32, return_sequences=False),
    Dense(2,activation='softmax')
])

model.summary()
```

```python
model.layers[0].input
```

```python
from tensorflow.keras.optimizers import Adam

# 새로운 학습률 설정
new_learning_rate = 0.0001

# 'adam' 옵티마이저를 생성하고 학습률 설정
optimizer = Adam(learning_rate=new_learning_rate)

# 모델 컴파일 시 새로운 옵티마이저를 사용
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
batch_size = 64
epochs = 40
from tensorflow.keras.callbacks import ModelCheckpoint

# ModelCheckpoint 콜백을 설정
model_checkpoint = ModelCheckpoint(f'{number}_model.h5', monitor='val_loss', save_best_only=True, mode='min', verbose=1)

# 모델 학습 시 ModelCheckpoint 콜백을 사용
history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    batch_size=batch_size,
                    epochs=epochs,
                    callbacks=[model_checkpoint])
```

```python
from tensorflow.keras.models import load_model

# 저장된 모델 파일을 불러오기
loaded_model = load_model(f'{number}_model.h5')
```
