```python
import pandas as pd
total1=pd.read_csv('./merge_body.csv')
```

```python
total2=pd.read_csv('./merge_barbell.csv')
```

```python
total = pd.concat([total1, total2], axis=0)
```

```python
total.loc[total['exercise'] == '바벨 로우', 'class'] = 1
total.loc[total['exercise'] == '푸시업', 'class'] = 4
total.loc[total['exercise'] == '오버 헤드 프레스', 'class'] = 3
total.loc[total['exercise'] == '바벨 데드리프트', 'class'] = 0
total.loc[total['exercise'] == '바벨 스쿼트', 'class'] = 2
```

```python
total
```

```python
import pandas as pd

# 데이터프레임을 32개씩 묶어서 리스트에 저장
sequence_length = 32
sequences = [total[i:i+sequence_length] for i in range(0, len(total), sequence_length)]

# 널값이 있는 시퀀스를 제외하고 새로운 리스트에 저장
cleaned_sequences = [seq for seq in sequences if not seq.isnull().any().any()]

# 새로운 데이터프레임으로 합치기
total_cleaned = pd.concat(cleaned_sequences)

# 제거한 결과 확인
total_cleaned
```

```python
df=total_cleaned.iloc[:,:99]
df.reset_index(drop=True,inplace=True)
```

```python
import math
def calculate_3d(point1, point2, point3):
    # Calculate the vectors between the left hand, elbow, and shoulder landmarks
    point_1_2 = [(point2[i] - point1[i]) for i in range(3)]
    point_2_3 = [(point2[i] - point3[i]) for i in range(3)]

    # Calculate the dot product and the magnitudes of the vectors
    dot_product = sum([point_1_2[i] * point_2_3[i] for i in range(3)])
    point_1_2_mag = math.sqrt(sum([coord**2 for coord in point_1_2]))
    point_2_3_mag = math.sqrt(sum([coord**2 for coord in point_2_3]))

    # Calculate the angle between the left hand, elbow, and shoulder landmarks in degrees
    angle = math.degrees(math.acos(dot_product / (point_1_2_mag * point_2_3_mag)))

    return angle
```

```python
point_groups = [
    ('LEFT_SHOULDER', 'LEFT_ELBOW', 'LEFT_WRIST'),
    ('LEFT_HIP', 'LEFT_SHOULDER', 'LEFT_ELBOW'),
    ('LEFT_KNEE', 'LEFT_HIP', 'LEFT_SHOULDER'),
    ('LEFT_ANKLE', 'LEFT_KNEE', 'LEFT_HIP'),
    ('RIGHT_SHOULDER', 'RIGHT_ELBOW', 'RIGHT_WRIST'),
    ('RIGHT_HIP', 'RIGHT_SHOULDER', 'RIGHT_ELBOW'),
    ('RIGHT_KNEE', 'RIGHT_HIP', 'RIGHT_SHOULDER'),
    ('RIGHT_ANKLE', 'RIGHT_KNEE', 'RIGHT_HIP')
]

# 열 이름 리스트 만들기
column_names = [f'{group[0]}_ANGLE' for group in point_groups]

# 각 그룹의 각도 계산 및 데이터프레임에 추가
for group, col_name in zip(point_groups, column_names):
    df[col_name] = df.apply(lambda row: calculate_3d(
        [row[f'{group[0]}.x'], row[f'{group[0]}.y'], row[f'{group[0]}.z']],
        [row[f'{group[1]}.x'], row[f'{group[1]}.y'], row[f'{group[1]}.z']],
        [row[f'{group[2]}.x'], row[f'{group[2]}.y'], row[f'{group[2]}.z']]
    ), axis=1)
```

```python
X=df
y=total_cleaned['class']
X.shape,y.shape
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
num_classes = len(np.unique(y))
y = keras.utils.to_categorical(y, num_classes)
y.shape
```

```python
# 데이터를 16개의 시퀀스로 묶습니다.
import numpy as np
sequence_length = 32  # 시퀀스 길이 설정
Xsequence, ysequence = [], []

for i in range(0, len(X) - sequence_length + 1, sequence_length):
    Xsequence.append(X[i:i+sequence_length])
    ysequence.append(y[i+sequence_length-1])

Xsequence = np.array(Xsequence)
ysequence = np.array(ysequence)
```

```python
Xsequence.shape,ysequence.shape
```

```python
from sklearn.model_selection import train_test_split

n = Xsequence.shape[0]
# 먼저 train과 test를 분리합니다.
X_train, X_test, y_train, y_test = train_test_split(Xsequence, ysequence, test_size=0.1, shuffle=True, random_state=42)
# 다음으로 train을 train과 validation으로 분리합니다.
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, shuffle=True, random_state=42)
y_train.shape,X_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense, BatchNormalization
from tensorflow.keras.models import Sequential

# Sequential 모델 정의
model = Sequential()
model.add(LSTM(128, dropout=0.2, input_shape=(sequence_length, 107), return_sequences=True))
BatchNormalization(),
model.add(LSTM(64))
model.add(Dense(5, activation='softmax'))
# 모델 요약
model.summary()
```

```python
model.layers[0].input
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
batch_size = 64
epochs = 20
history = model.fit(X_train, y_train,
                    validation_data=(X_val , y_val),  # 검증 세트 비율
                    batch_size=batch_size,
                    epochs=epochs)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(test_accuracy)
```
